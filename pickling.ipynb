```python
import os
import cv2
import numpy as np
import pickle
```

```python
img_folder = os.path.join(os.getcwd(),"train")
```

```python
images = []
labels = []
```

```python
for i in os.listdir(img_folder):
        for j in os.listdir(os.path.join(img_folder,i)):
            image_path = os.path.join(os.path.join(img_folder,i),j)
            image = cv2.imread(image_path)
            image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
            image = cv2.resize(image,(100,100))
            images.append(image)
            labels.append(i)
        print(i,"------------------- completed")
```

```python
len(labels)
```

```python
set(labels)
```

```python
images = np.array(images)
labels = np.array(labels)
```

```python
print(images.shape)
print(labels.shape)
```

```python
with open(os.path.join(os.getcwd(),"images.p"),"wb") as f:
    pickle.dump(images,f)
with open(os.path.join(os.getcwd(),"labels.p"),"wb") as f:
    pickle.dump(labels,f)
    
```
