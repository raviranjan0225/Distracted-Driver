```python
image = cv2.imread(r"C:\Users\ravir\OneDrive\Desktop\state-farm-distracted-driver-detection\test\img_58.jpg")
```

```python
plt.imshow(image)
plt.show()
```

```python
image = cv2.resize(image,(100,100))
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
image = cv2.equalizeHist(image)
image = image.reshape(1,100, 100, 1)
image = image/255
```

```python
image.shape
```

```python
import numpy as np
```

```python
i = np.argmax(model.predict(image.reshape(1,100,100,1)))
```

```python
if i == 0:
    print('Safe driving')
elif i == 1:
    print('Texting - right')
elif i == 2:
    print('Talking on the phone - right')
elif i == 3:
    print('Texting - left')
elif i == 4:
    print('Talking on the phone - left')
elif i == 5:
    print('Operating the radio')
elif i == 6:
    print('Drinking')
elif i == 7:
    print('Reaching behind')
elif i == 8:
    print('Hair and makeup')
elif i == 9:
    print('Talking to passenger')
```

# Prediction On Video

```python
cap = cv2.VideoCapture(r"C:\Users\ravir\OneDrive\Desktop\state-farm-distracted-driver-detection\input_video.mp4")
```

```python
ret , image = cap.read()
while ret:
    ret , frame = cap.read()
    image = frame
    image = cv2.resize(image,(100,100))
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    image = cv2.equalizeHist(image)
    image = image.reshape(1,100, 100, 1)
    image = image/255
    i = np.argmax(model.predict(image))
    l = ['Safe driving','Texting - right','Talking on the phone - right','Texting - left','Talking on the phone - left','Operating the radio','Drinking','Reaching behind','Hair and makeup','Talking to passenger']
    cv2.putText(frame,l[i],(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),thickness=2)
    cv2.imshow("win",frame)
    if cv2.waitKey(30) == ord("q"):
        break
cv2.destroyAllWindows()
    
    
```
